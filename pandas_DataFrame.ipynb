### DataFrame(데이터 프레임) = (Series + Series + ...) + Column
- 다변수(복합변수)


#### DataFrame 생성

```python
list = [3,4,5,6]
pd.DataFrame(list)
```

```python
# pd.Series(list)
```

```python
double_list = [
             [1000,'과자','2019-12-31','반품'],
             [2000,'음료','2020-03-02','정상'],
             [3000,'아이스크림','2020-02-03','정상'],
             [1000,'과자','2019-12-31','반품']
            ]
pd.DataFrame(data=double_list)
```

```python
double_columns = ['가격','종류','판매일자','반품여부']
pd.DataFrame(data=double_list,columns=double_columns)
```

#### DataFrame 내용 확인


```python
df_saledays = pd.DataFrame(data=double_list,columns=double_columns)
type(df_saledays), df_saledays.shape
```

### Row(Index) 단위
- iloc : 인덱스 번호 위치값 접근
- loc  : 이름으로 접근

```python
df_saledays
```

```python
df_saledays.index = ['a','b','c','d']
```

```python
# df_saledays.iloc[2],
df_saledays.loc['c']
```

### Index + Column 동시 추출

```python
df_saledays.loc['c', '판매일자'] # indexes, columns
```

```python
df_saledays.loc[ :, '판매일자']
```

```python
df_saledays.loc[:, '판매일자' :]
```

### 정렬

```python
df_saledays.sort_values(by='가격',ascending=True) # 단일 변수일 때 사용
```

```python
df_saledays.sort_values(by=['가격','반품여부'],ascending=[False,True])
```
